```python
import pandas as pd
import matplotlib.pyplot as plt

# Gi s df l DataFrame ban u ca bn

# Bc 1: Lc d liu cho quc gia MEX
mex_data = df[df['country_code'] == 'MEX']

# Bc 2: Nhm d liu theo 'disciplines' v m s lng ngi tham gia trong mi mn th thao
participant_count = mex_data.groupby('disciplines').size().reset_index(name='Number of Participants')

# Bc 3: V biu  ct (bar chart)
plt.figure(figsize=(15, 5))
bars=plt.bar(participant_count['disciplines'], participant_count['Number of Participants'], color='skyblue')

# t tiu  v nhn cho trc
plt.title('S lng ngi tham gia tng mn th thao ca MEX', fontsize=20)
plt.xlabel('Mn th thao', fontsize=16)
plt.ylabel('S lng ngi tham gia', fontsize=16)

# Xoay nhn trn trc x  d c hn
plt.xticks(rotation=45, ha='right')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, int(yval), ha='center', va='bottom', fontsize=12)
# Hin th biu 
plt.show()
```
