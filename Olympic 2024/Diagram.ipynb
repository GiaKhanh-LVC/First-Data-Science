```python
import pandas as pd
import matplotlib.pyplot as plt

# Giả sử df là DataFrame ban đầu của bạn

# Bước 1: Lọc dữ liệu cho quốc gia MEX
mex_data = df[df['country_code'] == 'MEX']

# Bước 2: Nhóm dữ liệu theo 'disciplines' và đếm số lượng người tham gia trong mỗi môn thể thao
participant_count = mex_data.groupby('disciplines').size().reset_index(name='Number of Participants')

# Bước 3: Vẽ biểu đồ cột (bar chart)
plt.figure(figsize=(15, 5))
bars=plt.bar(participant_count['disciplines'], participant_count['Number of Participants'], color='skyblue')

# Đặt tiêu đề và nhãn cho trục
plt.title('Số lượng người tham gia từng môn thể thao của MEX', fontsize=20)
plt.xlabel('Môn thể thao', fontsize=16)
plt.ylabel('Số lượng người tham gia', fontsize=16)

# Xoay nhãn trên trục x để dễ đọc hơn
plt.xticks(rotation=45, ha='right')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, int(yval), ha='center', va='bottom', fontsize=12)
# Hiển thị biểu đồ
plt.show()
```
