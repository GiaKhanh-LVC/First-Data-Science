```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df=pd.read_csv('athletes new.csv')
dt=df.drop(['code','name_tv','name','country','country_full','nationality','nationality_full','nationality_code'],axis=1)
```

```python
# Fill missing values in 'weight' with the median and convert to integer
dt['weight'].fillna(dt['weight'].median(), inplace=True)
dt['weight'] = dt['weight'].astype(int)

# Drop rows where 'height' or 'weight' is 0
dt = dt[(dt['height'] != 0) & (dt['weight'] != 0)]

# Print unique weight values
print(dt['weight'].unique())
print(dt['height'].unique())
dt.head()
```
